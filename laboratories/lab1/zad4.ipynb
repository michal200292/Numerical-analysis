```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eps = pow(10, -6)
def next_value(r, x):
    return r*x*(1 - x)
```

```python
def calculate_sequence(r, x):

    values = [x]
    for _ in range(pow(10, 5)):
        x = next_value(r, x)
        values.append(x)

    last = -1
    for i in range(len(values) - 2, len(values) - 20, -1):
        if abs(values[-1] - values[i]) < eps:
            last = i
            break
    if last > 0:
        return values[last+1:]
    return values[-50:]
```

```python
def plot_given_start_val(x0):
    step = 0.03
    err = 0
    r_val = 1.5
    x_axis = []
    y_axis = []
    while r_val < 4:
        new_points = calculate_sequence(r_val, x0)
        x_axis += len(new_points)*[r_val]
        y_axis += new_points
        y = step - err
        temp = r_val + y
        err = (temp - r_val) - y
        r_val = temp
    plt.scatter(x_axis, y_axis, s=5)
    plt.show()
```

```python
plot_given_start_val(0.1)
```

```python
plot_given_start_val(0.2)
```

```python
plot_given_start_val(0.5)
```

```python
def number_of_iter_to_reach_zero(x0):
    i = 0
    while abs(x0) > eps:
        x0 = next_value(4, x0)
        i += 1
    return i
```

```python
start = 0.05
x_ax = []
y_ax = []
while start < 1:
    x_ax.append(start)
    y_ax.append(number_of_iter_to_reach_zero(start))
    start += 0.02

plt.plot(x_ax, y_ax)
plt.show()
```
