```python
def riemman_sum_forwards(s, n):
    acc = 0
    for i in range(1, n + 1):
        acc += pow(i, -s)
    return acc

def riemman_sum_backwards(s, n):
    acc = 0
    for i in range(n, 0, -1):
        acc += pow(i, -s)
    return acc

def dirichlet_sum_forwards(s, n):
    sign = 1
    acc = 0
    for i in range(1, n + 1):
        acc += sign*pow(i, -s)
        sign *= -1
    return acc

def dirichlet_sum_backwards(s, n):
    sign = 1
    if n % 2 == 0:
        sign = -1
    acc = 0
    for i in range(n, 0, -1):
        acc += sign*pow(i, -s)
        sign *= -1
    return acc
```

```python
functions_to_cmp = [(riemman_sum_forwards, riemman_sum_backwards, "Riemman"),
                    (dirichlet_sum_forwards, dirichlet_sum_backwards, "Dirichlet")]

s_values = [2, 3.6667, 5, 7.2, 10]
n_values = [50, 100, 200, 500, 1000]

for a, b, c in functions_to_cmp:
    print(c, " sum compare")
    print("-----||-------")
    for s_value in s_values:
        for n_value in n_values:
            forward = a(s_value, n_value)
            backward = b(s_value, n_value)
            print("S =", s_value, "N =", n_value)
            print("Forward sum", forward)
            print("Backward sum", backward)
            print("Difference ", forward - backward)
            print()


```
